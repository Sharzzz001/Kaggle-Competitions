BMI, Menthlth, Physhlth have outliers

outlier removal

```python
def max_value(i, variable, top):
    return np.where(i[variable]>top, top, i[variable])

for i in [df1]:
    i['BMI'] = max_value(i, 'BMI', 57.0)
    i['MentHlth'] = max_value(i, 'MentHlth', 8.0)
    i['PhysHlth'] = max_value(i, 'PhysHlth', 24.0)
```

Train Test Split

```python
X = df1.drop(['Diabetes'],axis=1)
y = df1.Diabetes
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, 
                                                    random_state = 2)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
```

Use StandardScaler for normally distributed data, otherwise use MinMaxScaler.

```python
cols = X_train.columns
```

```python
scaler = MinMaxScaler()

X_train_LR = scaler.fit_transform(X_train)
X_test_LR = scaler.transform(X_test)
```

```python
X_train_LR = pd.DataFrame(X_train_LR,columns=cols)
X_test_LR = pd.DataFrame(X_test_LR,columns=cols)
```

```python
X_train_LR.describe()
```

Model Building

```python
from sklearn.linear_model import LogisticRegression


# instantiate the model
logreg = LogisticRegression(solver='liblinear', random_state=0)


# fit the model
logreg.fit(X_train_LR, y_train)
```

```python
# Create the predictions
#
y_pred= logreg.predict(X_test_LR)
 
# Use metrics.accuracy_score to measure the score
print("LogisticRegression Accuracy %.3f" %metrics.accuracy_score(y_test, y_pred))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(logreg.score(X_train_LR, y_train)))

print('Test set score: {:.4f}'.format(logreg.score(X_test_LR, y_test)))
```

These two values are quite comparable. So, there is no question of overfitting.

Using gridsearch for maximum accuracy

```python
from sklearn.model_selection import GridSearchCV


parameters = [{'penalty':['l1','l2']}, {'solver':['liblinear','sag','saga']},
              {'C':[0.01,0.1,1, 10, 100, 1000]}]


grid_search = GridSearchCV(estimator = logreg,  
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 5,
                           verbose=1)


grid_search.fit(X_train_LR, y_train)
print("Best Parameters are:",grid_search.best_params_)
```

```python
print('Training set score: {:.4f}'.format(grid_search.score(X_train_LR, y_train)))

print('Test set score: {:.4f}'.format(grid_search.score(X_test_LR, y_test)))
```

```python
y_pred= grid_search.predict(X_test_LR)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = np.arange(1,50)
#29 because sqrt of len(X_train)
train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k, random_state=2)
    
    #Fit the model
    knn.fit(X_train, y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)
    
    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(X_test, y_test) 
```

```python
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```
